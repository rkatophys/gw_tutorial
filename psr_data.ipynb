# パルサーのデータを確認するだけのノートブック

パルサーの観測データの中から必要なデータを取り出すには、特殊なパッケージが必要なので少し面倒。よって、ここでは、パルサーが観測された時間と、パルサーの自転周期の揺らぎの二つをまとめたデータを用意した。

このデータは、昨年、NANOGravというグループが公開したデータをもとにしている。

```python
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 全てのパルサーの名前
psrnames=['B1855+09', 'B1937+21', 'B1953+29', 'J0023+0923', 'J0030+0451',
       'J0340+4130', 'J0406+3039', 'J0437-4715', 'J0509+0856',
       'J0557+1551', 'J0605+3757', 'J0610-2100', 'J0613-0200',
       'J0636+5128', 'J0645+5158', 'J0709+0458', 'J0740+6620',
       'J0931-1902', 'J1012+5307', 'J1012-4235', 'J1022+1001',
       'J1024-0719', 'J1125+7819', 'J1312+0051', 'J1453+1902',
       'J1455-3330', 'J1600-3053', 'J1614-2230', 'J1630+3734',
       'J1640+2224', 'J1643-1224', 'J1705-1903', 'J1713+0747',
       'J1719-1438', 'J1730-2304', 'J1738+0333', 'J1741+1351',
       'J1744-1134', 'J1745+1017', 'J1747-4036', 'J1751-2857',
       'J1802-2124', 'J1811-2405', 'J1832-0836', 'J1843-1113',
       'J1853+1303', 'J1903+0327', 'J1909-3744', 'J1910+1256',
       'J1911+1347', 'J1918-0642', 'J1923+2515', 'J1944+0907',
       'J1946+3417', 'J2010-1323', 'J2017+0603', 'J2033+1734',
       'J2043+1711', 'J2124-3358', 'J2145-0750', 'J2214+3000',
       'J2229+2643', 'J2234+0611', 'J2234+0944', 'J2302+4442',
       'J2317+1439', 'J2322+2057']
```

```python
# ./Data以下にある、パルサーのデータのロード
# パルサーの名前をkey、データをvalueとする辞書を作る。
# それぞれのパルサーのデータ(value)の配列の0番目は観測された時間、1番目は自転周期の揺らぎになっている。
data_dict={}
for psrname in psrnames:
    data=np.loadtxt(glob.glob(f"./Data/{psrname}*")[0])
    data_dict[psrname]=data
```

```python
# J1909-3744の観測された時間(TOAs)と自転周期の揺らぎ (Residuals)のデータ。
psrname='J1909-3744'
datax=data_dict[psrname][0]
datay=data_dict[psrname][1]

# プロット。
plt.figure(figsize=(10,5))
plt.plot(datax,datay,".")
plt.xlabel("TOAs [s]")
plt.ylabel("Residuals [s]")
plt.title(f"{psrname}")

print(f"データの数: {len(datax)}")
```

```python
# 他のパルサーのデータを見たい場合は、psrnameを対応する名前に変える。
psrname='J1713+0747'
datax=data_dict[psrname][0]
datay=data_dict[psrname][1]

# プロット。
plt.figure(figsize=(10,5))
plt.plot(datax,datay,".")
plt.xlabel("TOAs [s]")
plt.ylabel("Residuals [s]")
plt.title(f"{psrname}")

print(f"データの数: {len(datax)}")
```
